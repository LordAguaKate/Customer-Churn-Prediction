```python
import joblib
import pandas as pd
from pathlib import Path
```

```python
root_dir = Path("../")
model_path = root_dir / "models" / "modelo_churn_v1.pkl"
data_path = root_dir / "data" / "processed" / "teleco_churn_final.csv"
```

```python
# 2. Cargar el modelo entrenado (Pipeline)
try:
    model = joblib.load(model_path)
    print(f"¡Modelo importado exitosamente desde: {model_path}!")
    print("El modelo ya incluye el escalado y la regresión logística. Listo para usar.")
except FileNotFoundError:
    print("ALERTA: No se encontró el archivo .pkl.")
    print("Asegúrate de haber ejecutado el código de 'joblib.dump' al final del notebook 0.3")
```

```python
df_ejemplo = pd.read_csv(data_path)

cliente_nuevo = df_ejemplo.drop('Churn', axis=1).iloc[[25]]

print("--- Datos del Cliente a Evaluar ---")
print(cliente_nuevo.T) # .T lo transpone para leerlo mejor verticalmente

prediccion = model.predict(cliente_nuevo)
probabilidad = model.predict_proba(cliente_nuevo)

print("\n" + "="*30)
print(f"Predicción del Modelo: {'SE VA' if prediccion[0] == 1 else 'SE QUEDA'}")
print(f"Probabilidad de fuga: {probabilidad[0][1]:.2%}")
print("="*30)
```
